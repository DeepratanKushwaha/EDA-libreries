```python
df = pd.read_csv('D:\BCA- STUDY\Python\Datasets\googleplaystore.csv', nrows=1000)
df.shape
```

```python
df
```

```python
x = df['Rating']
y = df['Reviews']
z = df['Installs']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Google play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c = 'r', marker='^', s = 170, alpha=0.2, linewidths=10, edgecolors='m') # verts='>'
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.show()
```

```python
# ploting 2 scatter in graph

plt.figure(figsize=(16,9))

plt.scatter(x,y, c = 'r', marker='*', s = 170, alpha=0.2, linewidths=10, edgecolors='m') # verts='>'
plt.scatter(x,z, c = 'y', marker='o', s = 170, alpha=0.2, linewidths=10, edgecolors='c') # verts='>'

plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Rating')
plt.ylabel('Reviews & Installs')
plt.show()
```
