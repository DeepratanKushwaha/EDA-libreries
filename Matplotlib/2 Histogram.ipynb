# Ploting Histograms

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18, 45, (100))
py_student_age = np.random.randint(15, 40, (100))

print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9)
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='barstacked')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='step')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='stepfilled')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', align='left')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', align='right')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', orientation='horizontal')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', color='r')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', color='y')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', color='r', label='ML Student')
plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.legend(loc=2)
plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth=0.9, histtype='bar', color='y', label='ML Student')
plt.hist(py_student_age, bins, rwidth=0.9, histtype='bar', color='m', label='Py Student')

plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.legend(loc=2)

plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype='bar', color=['y', 'm'], 
         label=['ML Student', 'PY Student'])
# plt.hist(py_student_age, bins, rwidth=0.9, histtype='bar', color='m', label='Py Student')

plt.title('ML Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.legend(loc=2)

plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize=[16,9])
style.use('ggplot')
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype='bar', color=['y', 'm'], 
         label=['ML Student', 'PY Student'])

plt.title('ML & PY Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.legend(loc=2)

plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize=[16,9], facecolor='r')
style.use('ggplot')
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype='bar', color=['y', 'm'], 
         label=['ML Student', 'PY Student'])

plt.title('ML & PY Students age histogram')
plt.xlabel('Student age category')
plt.ylabel('No. Student age')
plt.legend(loc=2)

plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize=[16,9], edgecolor='m', facecolor='m')
style.use('ggplot')
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype='bar', color=['y', 'm'], 
         label=['ML Student', 'PY Student'])

plt.title('ML & PY Students age histogram')
plt.xlabel('Student age category', color='w')
plt.ylabel('No. Student age', color='w')
plt.legend(loc=2)

plt.show()
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize=[16,9], edgecolor='m')
style.use('ggplot')
plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype='bar', color=['y', 'm'], 
         label=['ML Student', 'PY Student'])

plt.title('ML & PY Students age histogram')
plt.xlabel('Student age category', color='w')
plt.ylabel('No. Student age', color='w')
plt.legend(loc=2)

plt.show()
```
