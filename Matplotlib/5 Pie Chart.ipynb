```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'R', 'Machine Learning', 'Artificial Intelligent', 'Data Science']
class1_students = [45, 15, 35, 25, 30]
```

```python
plt.pie([1,2,3,4,5])
```

```python
plt.pie(class1_students)
plt.show()
```

```python
plt.pie(class1_students, labels=classes)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=100)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=100, textprops = {'fontsize':15})
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=100, textprops = {'fontsize':15}, pctdistance=0.8, labeldistance=1.2)
plt.show()
```

```python

plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=270, textprops = {'fontsize':15, 'color':'k'}, pctdistance=0.8, 
        labeldistance=1.2, wedgeprops={'linewidth':2, 'width':3,'edgecolor':'k'}, center=(2,3), frame=True, 
       rotatelabels=True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=270, textprops = {'fontsize':15, 'color':'k'}, pctdistance=0.8, 
        labeldistance=1.2, wedgeprops={'linewidth':2, 'width':3,'edgecolor':'k'}, center=(2,3), frame=True, 
       rotatelabels=True)
plt.show()
```

```python
plt.pie(class1_students, labels=classes, explode=[0,0,0.1,0,0], colors=['c', 'b', 'r', 'y', 'g'], autopct='%0.2f%%', 
       shadow=True, radius=2, startangle=100, textprops = {'fontsize':15}, pctdistance=0.8, labeldistance=1.2)
plt.legend(loc=4)
plt.show()
```
